## Part 1 — The Basic `Point` Class

### What is a class?

A **class** is like a blueprint for creating objects. Think of it like a cookie cutter — the class is the cutter shape, and each cookie you make is an **object** (also called an **instance**).

We are going to create a `Point` class that represents a point on a 2D grid (like an x,y coordinate on a map).

### Step 1: The `__init__` method (the Constructor)

Every class needs a special method called `__init__`. This runs **automatically** when you create a new object.

- `self` refers to the object being created (think of it as "this object")
- `self.x` and `self.y` store the values **inside** the object — these are called **attributes**

```python
class Point:
    """Simple class to represent a point in 2D space."""

    def __init__(self, x, y):
        self.x = x  # x is an instance attribute
        self.y = y  # y is an instance attribute
```

### Creating objects from the class

Now let's **create** two Point objects and access their attributes:

```python
p1 = Point(1, 2)
p2 = Point(3, 4)
p3 = Point(y=2, x=-7)

print("p1 coordinates:", p1.x, p1.y)
print("p2 coordinates:", p2.x, p2.y)
print("p3 coordinates:", p3.x, p3.y)
```

### Step 2: The `__str__` method (how the object prints)

If we try `print(p1)` right now, we get something ugly like `<__main__.Point object at 0x...>`. That’s not helpful!

The `__str__` method lets us control what `print()` shows. Let's rebuild the class with this method:

```python
class Point:
    """Simple class to represent a point in 2D space."""

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self): # is used when printing an instance
        return f"<{self.x},{self.y}>"

    def __repr__(self): # this one is used when printing a list of instances
        return "Bob" + self.__str__()
```

> **Note:** `__repr__` is similar to `__str__` but is used when Python needs a representation of the object (for example, when you have a list of Points). Here we just make it call `__str__` so they show the same thing.

Let’s test it:

```python
p1 = Point(1, 2)
p2 = Point(3, 4)

print(p1)       # Uses __str__
print(p2)       # Uses __str__
print([p1, p2]) # Uses __repr__ for items in a list
```

### Step 3: Adding custom methods

Methods are just **functions that live inside a class**. They always take `self` as the first parameter so they can access the object’s attributes.

Let’s add two useful methods:
- `distance_origin()` — distance from the point to (0, 0)
- `distance_to(point)` — distance from this point to another point

```python
class Point:
    """Simple class to represent a point in 2D space."""

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"P<{self.x},{self.y}>"

    def __repr__(self):
        return self.__str__()

    def distance_origin(self):
        """Distance from this point to the origin (0,0)."""
        return (self.x ** 2 + self.y ** 2) ** 0.5

    def distance_to(self, point):
        """Distance from this point to another point."""
        return ((self.x - point.x) ** 2 + (self.y - point.y) ** 2) ** 0.5
```

```python
p3 = Point(12, 5)
print(f"{p3} distance to origin is {p3.distance_origin()}")

p1 = Point(6, 10)
p2 = Point(6, 15)
print(f"Distance between {p1} and {p2} is {p1.distance_to(p2)}")
```

### Step 4: The `__lt__` method (comparing objects)

What if we want to **sort** a list of points? Python doesn't know how to compare two Point objects by default. We need to tell it!

The `__lt__` method defines what **"less than"** means. Here, we say a point is "less than" another if it is **closer to the origin**.

```python
class Point:
    """Simple class to represent a point in 2D space."""

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"P<{self.x},{self.y}>"

    def __repr__(self):
        return self.__str__()

    def distance_origin(self):
        """Distance from this point to the origin (0,0)."""
        return (self.x ** 2 + self.y ** 2) ** 0.5

    def distance_to(self, point):
        """Distance from this point to another point."""
        return ((self.x - point.x) ** 2 + (self.y - point.y) ** 2) ** 0.5

    def __lt__(self, other):
        """A point is 'less than' another if it is closer to the origin."""
        return self.distance_origin() < other.distance_origin()
```

Now we can sort a list of points! Python uses `__lt__` behind the scenes when you call `.sort()`.

```python
p1 = Point(6, 10)
p2 = Point(6, 15)
p3 = Point(12, 5)
p4 = Point(1, 1)

points = [p1, p2, p3, p4, Point(15, 6)]
print("Before sorting:", points)

points.sort(reverse=True)
print("After sorting: ", points)
```

## Part 2 — Inheritance with `ColorPoint`

### What is inheritance?

Inheritance lets you create a **new class based on an existing one**. The new class (called the **child**) gets all the methods and attributes of the original class (called the **parent**) for free!

Think of it like this: a `ColorPoint` **is a** `Point`, but with an extra attribute — a color.

We write `class ColorPoint(Point):` to say *"ColorPoint inherits from Point"*.

```python
class ColorPoint(Point):
    """A point that also has a color."""

    def __init__(self, x, y, color):
        # Check that x and y are numbers
        if not isinstance(x, (int, float)):
            raise TypeError("x must be a number")
        if not isinstance(y, (int, float)):
            raise TypeError("y must be a number")

        self.x = x
        self.y = y
        self.color = color

    def __str__(self):
        """Override the parent's __str__ to also show the color."""
        return f"{self.color}: <{self.x},{self.y}>"
```

**Key things to notice:**

- `ColorPoint` has its own `__init__` that adds a `color` attribute
- `ColorPoint` **overrides** `__str__` to include the color in the output
- It still **inherits** `distance_origin()`, `distance_to()`, `__lt__`, and `__repr__` from `Point` — we get those for free!
- We also added **type checking** with `isinstance()` to make sure x and y are numbers

```python
p1 = ColorPoint(1, 2, "red")
p2 = ColorPoint(3, 4, "green")
p3 = ColorPoint(5, 6, "blue")
p4 = ColorPoint(-2, -3, "yellow")

print(p1)          # Shows the color!
print(p1.color)    # Access the color attribute directly
try:
    p2 = ColorPoint(7.1, "Bob", "red")
except TypeError:
    print("I am sorry but that is not a valid number for a point coordinate")
```

### Inherited methods still work!

Even though we didn't write `distance_origin` or `__lt__` in `ColorPoint`, they work because they were inherited from `Point`:

```python
print(f"Distance to origin of p1={p1.distance_origin()}")
color_points = [p1, p2, p3, p4]
print("Before sorting:", color_points)

color_points.sort()  # This uses __lt__ inherited from Point!
print("After sorting: ", color_points)
```

## Part 3 — Advanced Features with `AdvancePoint`

Now we’ll look at some more powerful Python class features:

- **Class variables** — shared by ALL objects of the class
- **Properties** (`@property`) — controlled access to attributes
- **Class methods** (`@classmethod`) — methods that belong to the class, not an instance
- **Static methods** (`@staticmethod`) — utility functions that live inside the class

### Class variables vs instance variables

- **Instance variables** (like `self.x`) belong to ONE specific object
- **Class variables** (like `COLORS`) are shared across ALL objects of that class

Think of it like this: every student (instance) has their own name, but the school rules (class variable) are the same for everyone.

```python
class AdvancePoint(ColorPoint):
    # Class variable — shared by ALL AdvancePoint objects
    COLORS = ["red", "green", "blue", "yellow", "black", "white"]

    def __init__(self, x, y, color):
        # Check the color is in the allowed list
        if color not in self.COLORS:
            raise ValueError(f"Invalid color: must be one of {self.COLORS}")
        self._x = x        # underscore = "private" (convention)
        self._y = y
        self._color = color
```

> **Why the underscore?** Writing `self._x` instead of `self.x` is a Python convention meaning *"this attribute is private — don’t access it directly from outside the class."* We use **properties** (next section) to provide safe access.

### Properties: Getters and Setters

A **property** lets you access an attribute like normal (`point.x`) but behind the scenes, a method runs. This is useful for:
- Making attributes **read-only**
- **Validating** values before they are set

Let’s rebuild the full class with properties:

```python
class AdvancePoint(ColorPoint):
    COLORS = ["red", "green", "blue", "yellow", "black", "white"]

    def __init__(self, x, y, color):
        if color not in self.COLORS:
            raise ValueError(f"Invalid color: must be one of {self.COLORS}")
        self._x = x
        self._y = y
        self._color = color

    # ----- Properties (Getters) -----
    @property
    def x(self):
        return self._x

    @property
    def y(self):
        return self._y

    @property
    def color(self):       # This is a GETTER
        return self._color

    # ----- Property Setter -----
    @color.setter
    def color(self, value):  # This is a SETTER
        if value not in self.COLORS:
            raise ValueError(f"Invalid color: must be one of {self.COLORS}")
        self._color = value
```

**What’s happening here?**

- `x` and `y` only have a **getter** → they are **read-only**
- `color` has both a **getter** and a **setter** → you can change it, but only to a valid color

Let’s see it in action:

```python
p1 = AdvancePoint(1, 2, "red")
print(p1)
print("x =", p1.x)  # Uses the @property getter

# Change the color (uses the setter, which validates!)
p1.color = "blue"
print("After changing color:", p1)
```

### Class methods (`@classmethod`)

A **class method** belongs to the class itself, not to any specific object. It uses `cls` instead of `self`.

This is useful for operations that affect the whole class — like adding a new allowed color:

```python
class AdvancePoint(ColorPoint):
    COLORS = ["red", "green", "blue", "yellow", "black", "white"]

    def __init__(self, x, y, color):
        if color not in self.COLORS:
            raise ValueError(f"Invalid color: must be one of {self.COLORS}")
        self._x = x
        self._y = y
        self._color = color

    @property
    def x(self):
        return self._x

    @property
    def y(self):
        return self._y

    @property
    def color(self):
        return self._color

    @color.setter
    def color(self, value):
        if value not in self.COLORS:
            raise ValueError(f"Invalid color: must be one of {self.COLORS}")
        self._color = value

    # ----- Class Method -----
    @classmethod
    def add_color(cls, color):
        """Add a new color to the allowed list for ALL AdvancePoint objects."""
        cls.COLORS.append(color)

    # ----- Static Methods -----
    @staticmethod
    def distance_2_points(p1, p2):
        """Calculate distance between any two points."""
        return ((p1.x - p2.x) ** 2 + (p1.y - p2.y) ** 2) ** 0.5

    @staticmethod
    def from_dict(d):
        """Factory method: create an AdvancePoint from a dictionary."""
        return AdvancePoint(d["x"], d["y"], d["color"])
```

### Using `@classmethod` — adding a new color

Notice we call `add_color` on the **class name** (`AdvancePoint.add_color(...)`) not on an instance:

```python
# "coral" is not in the COLORS list yet, so this would fail:
# p_fail = AdvancePoint(1, 2, "coral")  # ValueError!

# Add "coral" to the allowed colors for ALL AdvancePoint objects
AdvancePoint.add_color("coral")

# Now it works!
p3 = AdvancePoint(1, 2, "coral")
print(p3)
```

### Using `@staticmethod` — utility functions

A **static method** doesn’t use `self` or `cls` — it’s just a regular function that lives inside the class for organization.

We have two static methods:
- `distance_2_points(p1, p2)` — calculates distance between two points
- `from_dict(d)` — a **factory method** that creates an AdvancePoint from a dictionary

```python
p1 = AdvancePoint(1, 2, "red")
p2 = AdvancePoint(3, 4, "white")

# Static method called on the class name
dist = AdvancePoint.distance_2_points(p1, p2)
print(f"Distance between {p1} and {p2} is {dist}")
```

```python
# Factory method: create a point from a dictionary
point_dict = {"x": 1, "y": 2, "color": "red"}
p0 = AdvancePoint.from_dict(point_dict)
print(p0)
```
