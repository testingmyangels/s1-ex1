```python
import matplotlib.pyplot as plt
import numpy as np

# Data
flavors = ["Vanilla", "Chocolate", "Strawberry", "Mint Choc", "Cookie Dough"]
months  = ["June", "July", "August"]
sales   = np.array([
    [120, 135, 110],
    [200, 220, 190],
    [ 80,  95,  70],
    [ 60,  75,  85],
    [150, 175, 160],
])
colors = ["#f9c74f", "#6d4c41", "#f4a5ae", "#80b192", "#c77dff"]

# Create 3D plot
ax = plt.figure(figsize=(12, 7)).add_subplot(111, projection="3d")

# Draw bars
for i in range(len(flavors)):
    for j in range(len(months)):
        ax.bar3d(i, j, 0, 0.6, 0.6, sales[i, j], color=colors[i], alpha=0.85)

# Labels
ax.set_title("Ice Cream Sales by Flavor & Month")
ax.set_xticks(np.arange(len(flavors)) + 0.3); ax.set_xticklabels(flavors)
ax.set_yticks(np.arange(len(months))  + 0.3); ax.set_yticklabels(months)
ax.set_zlabel("Scoops Sold")
ax.view_init(elev=25, azim=-50)

plt.show()
```
