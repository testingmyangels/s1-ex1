## What is Pandas?
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.<br>

##Library Highlights
- A fast and efficient DataFrame object for data manipulation with integrated indexing;

- Tools for reading and writing data between in-memory data structures and different formats: CSV and text files, Microsoft Excel, SQL databases, and the fast HDF5 format;

- Intelligent data alignment and integrated handling of missing data: gain automatic label-based alignment in computations and easily manipulate messy data into an orderly form;

- Flexible reshaping and pivoting of data sets;

- Intelligent label-based slicing, fancy indexing, and subsetting of large data sets;

- Columns can be inserted and deleted from data structures for size mutability;

- Aggregating or transforming data with a powerful group by engine allowing split-apply-combine operations on data sets;

- High performance merging and joining of data sets;

- Hierarchical axis indexing provides an intuitive way of working with high-dimensional data in a lower-dimensional data structure;

````python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

# Fetch AAPL data
df_stock = yf.download('AAPL', start='2007-01-01', end='2026-04-20')

df_launch = pd.read_excel("sample_data/launch.xlsx", engine="openpyxl")

df_launch['Date'] = pd.to_datetime(df_launch.date)
df_stock['Date'] = pd.to_datetime(df_stock.index)
df_stock.insert(0, 'ID', range(0, len(df_stock)))
print(df_stock)
print(df_launch)

# Initialize list to store matched IDs
matched_ids = []

# Loop through each launch date
for launch_date in df_launch['Date']:
    # Try matching exact date, +1 day, or +2 days
    for offset in [0, 1, 2]:
        check_date = launch_date + pd.Timedelta(days=offset)
        matching_row = df_stock[df_stock['Date'] == check_date]

        # If a match is found, store the ID and stop checking offsets
        if not matching_row.empty:
            matched_ids.append(int(matching_row['ID'].iloc[0]))
            break
    else:
        # If no match is found after checking all offsets, print a message
        print(f"No trading day found for {launch_date} within 2 days")

print(matched_ids)

plt.figure("Apple Stock")
plt.figure(figsize = (20, 10))
plt.plot(df_stock["Date"], df_stock["Close"], '--v', ms=20, mfc="black", linewidth=0.6,
         markevery=matched_ids, label="Iphone launch date")
plt.xlabel("Dates")
plt.legend(loc="upper left")
plt.show()
df_stock.to_csv("sample_data/result.csv")
````

````python
# Plotting using pandas' built-in functionality
ax = df_stock['Close'].plot(
    figsize=(20, 10),
    style='-ob',
    linewidth=0.2,
    markevery=matched_ids,  # Highlight the launch dates
    markerfacecolor='#ffaa88',
    markersize=20,
    label='Close Price'
)

# Highlight iPhone launch dates with vertical lines
for launch_date in df_stock.iloc[matched_ids].index:
    ax.axvline(x=launch_date, color='gray', linestyle='--', linewidth=1.5)

# Labeling and legend
ax.set_xlabel("Date")
ax.set_ylabel("Close Price")
ax.set_title("Apple Stock Prices with iPhone Launch Dates Highlighted")
ax.legend(loc="upper left")

# Show the plot
plt.show()
````

# 🦠 COVID-19 Data Analysis with Pandas

## What we'll learn in this notebook:

### ⚠️ Background: The CDC changed their data source!

The old way to get COVID data used to be:
```python
# ❌ THIS NO LONGER WORKS (returns 404 error)
df = pd.read_json("https://covid.cdc.gov/covid-data-tracker/COVIDData/getAjaxData?id=us_trend_by_USA_v2")
```

The CDC retired the old COVID Data Tracker API. The data now lives on **data.cdc.gov** using the **Socrata Open Data API (SODA)**.

The dataset we'll use: **[Provisional COVID-19 Death Counts by Week Ending Date and State](https://data.cdc.gov/d/r8kw-7aab)**  
Source: National Center for Health Statistics (NCHS)

---
## Step 1: Imports

We only need three libraries — all pre-installed in Google Colab.

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

print(f"Pandas version: {pd.__version__}")
print("Ready to go! ✅")
````

---
## Step 2: Fetch COVID-19 Data from the CDC API

The new CDC data lives on `data.cdc.gov` and uses the **Socrata API**.

**Key things about the API URL:**
- `r8kw-7aab` is the **dataset ID** (like a unique code for this specific dataset)
- `$where` filters the data (we want US national totals, grouped by week)
- `$order` sorts chronologically
- `$limit` controls how many rows we get (50000 is more than enough)
- `` `group` `` needs backticks because "group" is a reserved word in SoQL

💡 **SoQL** = Socrata Query Language (similar to SQL)

````python
# Build the API URL
BASE_URL = "https://data.cdc.gov/resource/r8kw-7aab.json"

params = {
    "$where":  "state='United States' AND `group`='By Week'",
    "$order":  "end_date ASC",
    "$limit":  "50000"
}

# Pandas can read JSON directly from a URL!
from urllib.parse import urlencode
url = f"{BASE_URL}?{urlencode(params)}"
print("Fetching data from:")
print(url)
print("\nPlease wait...")

df = pd.read_json(url)
print(f"\n✅ Downloaded {len(df)} rows of data!")
df.head()
````

## Step 3: Explore the raw data

Before doing anything, let's understand what we have.

````python
# What columns do we have?
print("Columns:")
print(df.columns.tolist())
print(f"\nShape: {df.shape[0]} rows x {df.shape[1]} columns")
print(f"\nData types:")
print(df.dtypes)
````

````python
# Let's look at the first and last few rows
print("=== First 3 rows ===")
print(df.head(3))
print("\n=== Last 3 rows ===")
print(df.tail(3))
````

## Step 4: Clean and prepare the data

We need to:
1. Keep only the columns we care about
2. Set `week_ending_date` as our index
3. Make sure dates are proper `datetime` objects
4. Make sure numeric columns are actually numbers

````python
# Select only the columns we need
columns_to_keep = [
    'week_ending_date',
    'covid_19_deaths',
    'total_deaths',
    'pneumonia_deaths',
    'influenza_deaths',
    'pneumonia_and_covid_19_deaths'
]

# Only keep columns that exist (some might be missing)
columns_to_keep = [c for c in columns_to_keep if c in df.columns] # double check
df2 = df[columns_to_keep].copy()

print(f"Kept {len(columns_to_keep)} columns:")
print(columns_to_keep)
df2
````

````python
# Convert week_ending_date to datetime and set as index
df2['week_ending_date'] = pd.to_datetime(df2['week_ending_date'])
df3 = df2.set_index('week_ending_date')

# Make sure all numeric columns are actually numeric
for col in df3.columns:
    df3[col] = pd.to_numeric(df3[col], errors='coerce')

print("Index type:", type(df3.index))
print("\nData types after cleaning:")
print(df3.dtypes)
print(f"\nDate range: {df3.index.min()} to {df3.index.max()}")
df3.head()
````

````python
# Let's check the index
print("Index:")
print(df3.index)
print(f"\nTotal weeks of data: {len(df3)}")
````

---
## Step 5: Add a cumulative deaths column

The old CDC API had a `COVID_deaths_total` column (cumulative).
The new API only gives us **weekly** counts, so we compute the cumulative sum ourselves using `.cumsum()`.

````python
# Create cumulative (running total) column
df3['covid_deaths_cumulative'] = df3['covid_19_deaths'].cumsum()

print(f"Total cumulative COVID-19 deaths: {df3['covid_deaths_cumulative'].iloc[-1]:,.0f}")
print(f"Peak weekly deaths: {df3['covid_19_deaths'].max():,.0f} (week of {df3['covid_19_deaths'].idxmax().date()})")
df3[['covid_19_deaths', 'covid_deaths_cumulative']].tail(10)
````

## Step 10: COVID deaths as a percentage of all deaths

We can create new columns by doing math on existing ones.

The same API lets you filter by **state**. Let's compare a few states.

````python
def get_state_data(state_name):
    """Fetch weekly COVID death data for a specific state."""
    params = {
        "$where":  f"state='{state_name}' AND `group`='By Week'",
        "$order":  "end_date ASC",
        "$limit":  "50000",
        "$select": "week_ending_date,covid_19_deaths,total_deaths"
    }
    url = f"{BASE_URL}?{urlencode(params)}"
    state_df = pd.read_json(url)
    state_df['week_ending_date'] = pd.to_datetime(state_df['week_ending_date'])
    state_df = state_df.set_index('week_ending_date')
    state_df['covid_19_deaths'] = pd.to_numeric(state_df['covid_19_deaths'], errors='coerce')
    return state_df

# Fetch data for a few states
states = ['California', 'Texas', 'New York', 'Florida']
state_data = {}

for state in states:
    print(f"Fetching {state}...")
    state_data[state] = get_state_data(state)

print("\nDone! ✅")
````

````python
# Plot all states on the same chart
fig, ax = plt.subplots(figsize=(25, 10))

for state_name, state_df in state_data.items():
    ax.plot(state_df.index, state_df['covid_19_deaths'], linewidth=1.5, label=state_name)

ax.set_title('Weekly COVID-19 Deaths by State', fontsize=20, fontweight='bold')
ax.set_ylabel('Weekly Deaths', fontsize=14)
ax.set_xlabel('Date', fontsize=14)
ax.legend(fontsize=14)
ax.ticklabel_format(style='plain', axis='y')
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45, ha='right')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````
